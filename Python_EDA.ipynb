```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding = 'unicode escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.describe()
```

```python
df[['Age','Orders']].describe()
```

# EDA

```python
df.columns

```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
sales_gen
```

```python
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)    
```

From above graph, we can say that most of the buyers are female and female spend more amount than male.

# Age Group

```python
df.columns
```

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

From the above graphs, we can see that most of the buyers are Females from Age Group 26-35.

# Total No. of Orders from Top 10 States

```python
df.columns
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', data=sales_state)
```

# Total Amount of Sales from Top 10 States

```python
sales_amount = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State', y='Amount',data=sales_amount)
```

From above Graphs, we can see that most of the Orders and high total Amount of Sales is from U.P, Maharashtra and Karnataka. 

# Marital Status

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(5,5)})
ax = sns.countplot(x='Marital_Status', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital_status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Marital_Status', y='Amount',hue='Gender',data=sales_marital_status)
```

From the Above Graphs,we can see that Most of the Buyers who spend More Amount are Married Women.

# Occupation

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Occupation', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Occupation', y='Amount', data=sales_occupation)
```

From above Graphs, We can see that Most Buyers are from IT Sector,Healthcare and Aviation.

# Product Category

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(22,5)})
ax = sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category', y='Amount',data=sales_product)
```

From above graphs, we can see that Most Sold Products are from Food,Clothing and Electronics category.

# Orders by Product ID

```python
sales_order = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(['Orders'],ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Product_ID', y='Orders',data=sales_order)
```

From the above Graph, we can see that the Most Sold Product is Product ID: P00265242
