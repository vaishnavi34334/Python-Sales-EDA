```python
#Calulate the mean for each category and fill NA wih mean for that category
df['Amount']=df.groupby('Product_Category')['Amount'].transform(lambda x: x.fillna(x.mean()))
```

```python
#Categorizing age
def Categorized_age (age):
    if age<=17:
        return '0-17'
    elif age<=25:
        return '18-25'
    elif age<=35:
        return '26-35'
    elif age<=45:
        return '35-45'
    elif age<=50:
        return '46-50'
    elif age<=55:
        return '51-55'
    else:
        return '55+'
 #Apply the function to create new column "Age_group"
df['Age_Group']=df['Age'].apply(Categorized_age)
print(df)
```

```python
ax=sns.countplot(x='Gender',data=df)
#Adding label to show the count of each gender
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_gen=df.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=Sales_gen)
```

```python
Sales_gen=df.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=Sales_gen)
for bars in Sales_gen.containers:
     Sales_gen.bar_label(bars)    
```

```python
#Top 10 states placed the highest Orders
Sales_State=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=Sales_State,x='State',y='Orders')
```

```python
#Top 5 states placed with highest amount
Sales_State=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=Sales_State,x='State',y='Amount')
```
